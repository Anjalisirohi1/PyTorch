```python
import nltk
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
tokens = word_tokenize(document.lower())
```

```python
vocab = {'<UNK>':0}
for token in Counter(tokens).keys():
    if token not in vocab:
        vocab[token] = len(vocab)

vocab
```

```python
len(vocab)
```

```python
input_sentences = document.split('\n')
```

```python
input_sentences
```

```python
def text_to_indices(sentence,voccab):
    numerical_sentence = []

    for token in sentence:

        if token in vocab:
            numerical_sentence.append(vocab[token])

        else:
            numerical_sentence.append(vocab['<UNK>'])

    return numerical_sentence
```

```python
input_numerical_sentences = []

for sentence in input_sentences:
    input_numerical_sentences.append(text_to_indices(word_tokenize(sentence.lower()),vocab))
```

```python
len(input_numerical_sentences)
```

```python
training_Sequence = []

for sentence in input_numerical_sentences:

    for i in range(1,len(sentence)):
        training_Sequence.append(sentence[:i+1])
    
```

```python
training_Sequence[:5]
```

```python
len_list = []
for seq in training_Sequence:
    len_list.append(len(seq))

max(len_list)
```

```python
padded_training_seq = []
for seq in training_Sequence:
    padded_training_seq.append([0]*(max(len_list)-len(seq)) + seq)
```

```python
len(padded_training_seq[0])
```

```python
padded_training_seq = torch.tensor(padded_training_seq,dtype=torch.long)
```

```python
padded_training_seq
```

```python
X = padded_training_seq[:,:-1]
y = padded_training_seq[:,-1]
```

```python
X
```

```python
y
```

```python
class Custom_dataset(Dataset):

    def __init__(self,X,y):
        self.X = X
        self.y = y

    def __len__(self):
        return self.X.shape[0]
    
    def __getitem__(self, index):
        return self.X[index],self.y[index]
```

```python
dataset = Custom_dataset(X,y)
```

```python
len(dataset)
```

```python
datasloader = DataLoader(dataset,batch_size=32,shuffle=True)
```

```python
for ip,op in datasloader:
    print(ip,op)
```

```python
class LSTM_Model(nn.Module):

    def __init__(self,vocab_size):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size,100)
        self.lstm = nn.LSTM(100,150,batch_first=True)
        self.fc = nn.Linear(150,vocab_size)

    def forward(self,x):
        embedded = self.embedding(x)
        intermediate_hidden_states,(final_hidden_state,final_cell_state) = self.lstm(embedded)
        output = self.fc(final_hidden_state).squeeze(0)
        return output
        
```

```python
model = LSTM_Model(len(vocab))
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

```python
model.to(device)
```

```python
epochs = 50
learning_rate = 0.001

loss_fun = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)
```

```python
for epoch in range(epochs):

    total_loss = 0

    for batch_x,batch_y in datasloader:
        batch_x ,batch_y = batch_x.to(device),batch_y.to(device)

        optimizer.zero_grad()

        y_pred = model(batch_x)

        loss = loss_fun(y_pred,batch_y)

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch: {epoch+1},Loss: {total_loss:.4f}")
    
```

```python
def prediction(model,vocab,text):
    
    #tokenize
    tokenized_text = word_tokenize(text.lower())

    #text to indices
    numerical_text = text_to_indices(tokenized_text,vocab)

    #padding
    padded_text = torch.tensor([0]*(61-len(numerical_text)) + numerical_text,dtype=torch.long).unsqueeze(0).to(device)

    output = model(padded_text)

    value,index = torch.max(output,dim=1)

    return text + " " + list(vocab.keys())[index]
```

```python
prediction(model,vocab,"total duration of")
```
